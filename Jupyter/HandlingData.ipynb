```python
#import the library
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# load the dataset
dataset = pd.read_csv("../data/owid-covid-data.csv") 
```

```python
usefulCol = ['iso_code','continent', 'location', 'date', 'total_vaccinations','people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations','new_vaccinations_smoothed']
useddata = pd.DataFrame(dataset, columns= usefulCol)
useddata.head(5)
```

```python
continentName = ['Africa', 'Asia', 'Europe', 'Oceania', 'North America', 'South America']
continentIndex = []
countryIndex = []
for index, value in useddata.iterrows():
#     if value['continent'] not in continentName:
    if len(str(value['continent'])) < 4:
        continentIndex.append(index)
#         print(value['continent'])
    else:
        countryIndex.append(index)
continent = useddata.loc[continentIndex,:]
country = useddata.loc[countryIndex,:]
```

```python
continent_f = continent.iloc[:,2:]
continent_f
```

```python
continent_f1 = continent_f.dropna(how='all', subset=['total_vaccinations','people_vaccinated', 'people_fully_vaccinated', 'new_vaccinations','new_vaccinations_smoothed'])
continent_f2 = continent_f1.fillna(0)
continent_f2
```

```python
changeIndex = []
location = ""
for index, value in continent_f2.iterrows():
    if value["location"] != location:
        changeIndex.append(index)
        location = value["location"]
changeIndex    
```

```python
continent_f2.loc[changeIndex[0]:(changeIndex[1]-1), :].to_csv("../data/continent/Africa.csv", index=False)
continent_f2.loc[changeIndex[1]:(changeIndex[2]-1), :].to_csv("../data/continent/Asia.csv", index=False)
continent_f2.loc[changeIndex[2]:(changeIndex[3]-1), :].to_csv("../data/continent/Europe.csv", index=False)
continent_f2.loc[changeIndex[3]:(changeIndex[4]-1), :].to_csv("../data/continent/European Union.csv", index=False)
continent_f2.loc[changeIndex[4]:(changeIndex[5]-1), :].to_csv("../data/continent/NorthAmer.csv", index=False)
continent_f2.loc[changeIndex[5]:(changeIndex[6]-1), :].to_csv("../data/continent/Oceania.csv", index=False)
continent_f2.loc[changeIndex[6]:(changeIndex[7]-1), :].to_csv("../data/continent/SouthAmer.csv", index=False)
continent_f2.loc[changeIndex[7]:, :].to_csv("../data/continent/World.csv", index=False)
```

```python
# sorted(country_f1.date)
country_f1.sort_values('date', inplace=True)
```

```python
country_f1
```

```python
country_f2 = pd.concat([country_f1, pd.DataFrame(columns=['month'])], sort= True)
# fix the object in the season col.
for index, value in country_f2.iterrows():
    month = int(value.date.strftime('%m'))
    country_f2.loc[index, 'month'] = month
country_f2
```

```python
datepoint = ['2020-12-02', '2020-12-31' , '2021-01-31'
            , '2021-02-28', '2021-03-31'
            , '2021-04-30', '2021-05-31',"2021-06-01"]
```

```python
# changeDateIndex = []
indexN = 0
dateValue = pd.DataFrame( columns= country_f2.columns.to_list())
for index, value in country_f2.iterrows():
#     print(str(value['date']))
    date = str(value['date']).split(" ")[0]
#     print(date)
    if date in datepoint:
        indexN +=1
#         changeDateIndex.append(index)
        dateValue.loc[indexN, :] = value
# changeDateIndex
```
